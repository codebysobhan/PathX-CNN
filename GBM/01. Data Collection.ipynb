# Pathways

```python
pathway_m=  pd.read_excel("../MingonPaper_Items/ordered_pathway_146_2pc (2).xlsx", header=None)
pathway_m
```

```python
keggPathways= pd.read_excel("../01. Data Collection/c2.cp.kegg_legacy.v2023.2.Hs.symbols (1).gmt.xlsx", header=None)
keggPathways.index= keggPathways[0]
keggPathways= keggPathways.drop([0,1], axis='columns')
keggPathways
```

```python
pathwayName= []
genelist= []
geneNumber= []
for i in range(0, keggPathways.shape[0]):
    pathwayName.append(keggPathways.index[i][5:])
    genes= keggPathways.iloc[i,:].tolist()
    genes = [x for x in genes if str(x) != 'nan']
    genelist.append(genes)
    geneNumber.append(len(genes))
    
pathwaydf= pd.DataFrame()
pathwaydf['PathwayName']= pathwayName
pathwaydf['Genes']= genelist
pathwaydf['#Genes']= geneNumber
pathwaydf
```

```python
excluded= pd.DataFrame(set(pathwaydf['PathwayName'])- set(pathway_m[0]))
excluded.to_csv("01. Data Collection/excludedPathways.csv", header=None, index=False)
```

```python
# finding common genes among mingon list and kegg list

pathway_final= pd.merge(pathwaydf, pathway_m, left_on='PathwayName', right_on=0)
pathway_final= pathway_final.drop([0], axis='columns')
pathway_final
```

```python
pathway_final.to_csv("01. Data Collection/final_pathway.csv")
```
