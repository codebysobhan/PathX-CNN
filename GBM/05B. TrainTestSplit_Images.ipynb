```python
import os
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from shutil import copyfile

# Define folder paths and classes
directory= '05. Convert Images/'

folder_paths = [directory+'0', directory+'1']
class_labels = [0, 1]  # Assign numerical labels to classes
```

```python
# Initialize lists to store file paths and corresponding class labels
file_paths = []
labels = []

# Iterate through each folder
for class_label, folder_path in zip(class_labels, folder_paths):
    files = os.listdir(folder_path)
    for file in files:
        file_paths.append(os.path.join(folder_path, file))
        labels.append(class_label)

# Create DataFrame to hold file paths and labels
data = pd.DataFrame({'file_path': file_paths, 'label': labels})

# Initialize StratifiedKFold
num_folds = 5
skf = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

# Create directories for storing training and testing files
os.makedirs(directory+'Train_images', exist_ok=True)
os.makedirs(directory+'Test_images', exist_ok=True)

os.makedirs(directory+'Train_images/FileInfo', exist_ok=True)
os.makedirs(directory+'Test_images/FileInfo', exist_ok=True)

```

```python
# Split data into folds
for fold_idx, (train_index, test_index) in enumerate(skf.split(data['file_path'], data['label'])):
    train_data = data.iloc[train_index]
    test_data = data.iloc[test_index]
    
    # Create folders for training and testing data for this fold
    train_folder = os.path.join(directory+'Train_images', f'fold_{fold_idx + 1}')
    test_folder = os.path.join(directory+'Test_images', f'fold_{fold_idx + 1}')
    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(test_folder, exist_ok=True)
    
    # Copy training files to the train folder
    for idx, row in train_data.iterrows():
        src_file = row['file_path']
        class_name = src_file.split('/')[1].split('_')[0]
        class_name= class_name.split('\\')[0]
        dst_file = os.path.join(train_folder, f"{class_name}_{os.path.basename(src_file)}")
        copyfile(src_file, dst_file)
        train_data.to_csv(directory+'Train_images/FileInfo/Fold_' + str(fold_idx + 1) + "_File_Info.csv")
        
    # Copy testing files to the test folder
    for idx, row in test_data.iterrows():
        src_file = row['file_path']
        class_name = src_file.split('/')[1].split('_')[0]
        class_name= class_name.split('\\')[0]
        dst_file = os.path.join(test_folder, f"{class_name}_{os.path.basename(src_file)}")
        copyfile(src_file, dst_file)
        test_data.to_csv(directory+'Test_images/FileInfo/Fold_' + str(fold_idx + 1) + "_File_Info.csv")
        
    # Print fold details
    print(f"Fold {fold_idx + 1}:")
    print(f"Train samples: {len(train_data)}, Test samples: {len(test_data)}")
```
