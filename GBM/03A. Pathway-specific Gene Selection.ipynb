```python
subtypes= [0,1]
omics= ['CNV', 'EXP', 'MET']

# creating folders based on subtype name

def create_folders_from_list(folder_names):
    for name in folder_names:
        print(name)
        folder_path = os.path.join("03. Pathway-specific Gene Selection", str(name))
        os.makedirs(folder_path, exist_ok=True)

# Example usage
folder_names = subtypes
create_folders_from_list(folder_names)
```

```python
# read pathway file

pathwayfinal= pd.read_csv("01. Data Collection/final_pathway.csv")
pathwayfinal= pathwayfinal.drop(['Unnamed: 0'], axis='columns')

# Remove square brackets and single quotes from the specified column
pathwayfinal['Genes'] = pathwayfinal['Genes'].apply(lambda x: x.replace('[', ''))
pathwayfinal['Genes'] = pathwayfinal['Genes'].apply(lambda x: x.replace(']', ''))
pathwayfinal['Genes'] = pathwayfinal['Genes'].apply(lambda x: x.replace("'", ''))
pathwayfinal
```

```python
missing_allgenes_cnv=[]
missing_allgenes_exp=[]
missing_allgenes_met=[]
missing_allgenes_mut=[]
num_missing_allgenes_cnv= []
num_missing_allgenes_exp= []
num_missing_allgenes_met= []
num_missing_allgenes_mut= []

for i in subtypes:
    missing_genes_cnv= []
    num_missing_genes_cnv= []
    missing_genes_exp= []
    num_missing_genes_exp= []
    missing_genes_met= []
    num_missing_genes_met= []
    missing_genes_mut= []
    num_missing_genes_mut= []
    
    cnv_sub= pd.read_csv("02. Class-specific Sample/CNV_"+str(i) +".csv")
    exp_sub= pd.read_csv("02. Class-specific Sample/EXP_"+str(i) +".csv")
    met_sub= pd.read_csv("02. Class-specific Sample/MET_"+str(i) +".csv")
#     mut_sub= pd.read_csv("02. Class-specific Sample/MUT_"+i +".csv")
    for j in range(0, pathwayfinal.shape[0]):
        pathway_genes= pathwayfinal.iloc[j,1].split(', ') # created list of with name of pathway genes
        
        cnv_sub_path= cnv_sub
        cnv_sub_path.index= cnv_sub_path['Unnamed: 0']
        cnv_genes= cnv_sub_path.columns.tolist()
        commongenes= list(set(cnv_genes) & set(pathway_genes))
        uncommongenes= list(set(pathway_genes) - set(cnv_genes))
        missing_genes_cnv.append(uncommongenes)
        num_missing_genes_cnv.append(len(uncommongenes))     
        cnv_sub_path= cnv_sub_path[commongenes]
        cnv_sub_path.to_csv("03. Pathway-specific Gene Selection/"+str(i)+"/CNV_"+str(pathwayfinal['PathwayName'][j])+".csv")
        
        exp_sub_path= exp_sub
        exp_sub_path.index= exp_sub_path['Unnamed: 0']
        exp_genes= exp_sub_path.columns.tolist()
        commongenes= list(set(exp_genes) & set(pathway_genes))
        uncommongenes= list(set(pathway_genes) - set(exp_genes))
        missing_genes_exp.append(uncommongenes)
        num_missing_genes_exp.append(len(uncommongenes))     
        exp_sub_path= exp_sub_path[commongenes]
        exp_sub_path.to_csv("03. Pathway-specific Gene Selection/"+str(i)+"/EXP_"+str(pathwayfinal['PathwayName'][j])+".csv")
        
        met_sub_path= met_sub
        met_sub_path.index= met_sub_path['Unnamed: 0']
        met_genes= met_sub_path.columns.tolist()
        commongenes= list(set(met_genes) & set(pathway_genes))
        uncommongenes= list(set(pathway_genes) - set(met_genes))
        missing_genes_met.append(uncommongenes)
        num_missing_genes_met.append(len(uncommongenes))     
        met_sub_path= met_sub_path[commongenes]
        met_sub_path.to_csv("03. Pathway-specific Gene Selection/"+str(i)+"/MET_"+str(pathwayfinal['PathwayName'][j])+".csv")
        
#         mut_sub_path= mut_sub
#         mut_sub_path.index= mut_sub_path['patient']
#         mut_genes= mut_sub_path.columns.tolist()
#         commongenes= list(set(mut_genes) & set(pathway_genes))
#         uncommongenes= list(set(pathway_genes) - set(mut_genes))
#         missing_genes_mut.append(uncommongenes)
#         num_missing_genes_mut.append(len(uncommongenes))     
#         mut_sub_path= mut_sub_path[commongenes]
#         mut_sub_path.to_csv("03. Pathway-specific Gene Selection/"+i+"/MUT_"+str(pathwayfinal['PathwayName'][j])+".csv")
      
    
    missing_allgenes_cnv.append(missing_genes_cnv)
    num_missing_allgenes_cnv.append(num_missing_genes_cnv)
    missing_allgenes_exp.append(missing_genes_exp)
    num_missing_allgenes_exp.append(num_missing_genes_exp)
    missing_allgenes_met.append(missing_genes_met)
    num_missing_allgenes_met.append(num_missing_genes_met)
#     missing_allgenes_mut.append(missing_genes_mut)
#     num_missing_allgenes_mut.append(num_missing_genes_mut)
```

```python
# Summary of missing genes and number of missing genes for each omics and class
subtypes= [0,1]
# omics= ['CNV', 'EXP', 'MET', 'MUT']

for i in range(0, len(subtypes)):
    x= pd.DataFrame(pathwayfinal['PathwayName'])
    x['missingGenes_CNV']= missing_allgenes_cnv[i]
    x['NumofMissingGenes_CNV']= num_missing_allgenes_cnv[i]
    x['missingGenes_EXP']= missing_allgenes_exp[i]
    x['NumofMissingGenes_EXP']= num_missing_allgenes_exp[i]
    x['missingGenes_MET']= missing_allgenes_met[i]
    x['NumofMissingGenes_MET']= num_missing_allgenes_met[i]
#     x['missingGenes_MUT']= missing_allgenes_mut[i]
#     x['NumofMissingGenes_MUT']= num_missing_allgenes_mut[i]
    
    x.to_csv("03. Pathway-specific Gene Selection/Summary_missingGenes_"+str(subtypes[i])+".csv")
    
    




```
