```python
# CNN model

def keras_cnn(X, y, loss_func="sparse_categorical_crossentropy", input_shape=(146, 6, 1), num_classes=2):
    model = Sequential()

    model.add(Conv2D(64, (2, 2), input_shape=(146, 6, 1), padding="same")) # Conv2D(32, (3, 3), input_shape=(X.shape[1], X.shape[2], 1), padding="same")
    model.add(BatchNormalization(axis=-1))
    model.add(Activation("relu"))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(16, (2, 2), padding="same")) # Conv2D(32, (3, 3), input_shape=(X.shape[1], X.shape[2], 1), padding="same")
    model.add(BatchNormalization(axis=-1))
    model.add(Activation("relu"))
    model.add(MaxPooling2D(pool_size=(3, 3)))

    model.add(Conv2D(128, (3, 3), padding="same")) # Conv2D(32, (3, 3), input_shape=(X.shape[1], X.shape[2], 1), padding="same")
    model.add(BatchNormalization(axis=-1))
    model.add(Activation("relu"))

    model.add(Conv2D(128, (2, 2), padding="same")) # Conv2D(32, (3, 3), input_shape=(X.shape[1], X.shape[2], 1), padding="same")
    model.add(BatchNormalization(axis=-1))
    model.add(Activation("relu"))

    model.add(Flatten())
    model.add(Dropout(0.6000000000000001))

    model.add(Dense(720))
    model.add(Activation("tanh"))
    model.add(BatchNormalization())
    model.add(Dropout(0.0))

    model.add(Dense(num_classes))
    model.add(Activation("softmax"))

    model.compile(
        loss=loss_func,
#         optimizer=RMSprop(lr=0.01, rho=0.9, decay=0.01),
        optimizer=Adam(learning_rate=0.0003069273045576997,decay=0.0905127409782462),
        metrics=["accuracy"],
    )

    return model


model = keras_cnn(X_train_shuffled, y_train_shuffled)
print("Model Initiated.")

# Train model beautifully
model.fit(
    X_train_shuffled, 
    y_train_shuffled, 
    batch_size=32, 
    epochs=100,
    verbose=1,
    validation_split=0.2, 
) 
```
